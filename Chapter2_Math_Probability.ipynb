# Conditional Probability and Bayes' Theorem  
  
Probability of an event A occurring given event B has occurred. It is typically expressed as P(A Given B) or P(A|B).  
  
The direction of the condition matters, example, "the probability of having cancer given you are a coffee drinker" P(Cancer|Coffee) is different from "the probaility of being a coffee drinker given you have cancer" p(Coffee|Cancer).  
  
If we are interested in studying whether coffee contributes to cancer, we really are interested in the first conditional probability: the probability of someone having cancer given they are a coffee drinker.  

P(Coffee|Cancer) = 0.85.  
P(Cancer|Coffee) = ?  

How to flip the condition? There is a powerful little formula called Bayes' Theorem.

P(A|B) = P(B|A) * P(A) / p(B)



```python
# Bayes' Theorem in Python

p_coffee_drinker = 0.65
p_cancer = 0.05
p_coffee_drinker_given_cancer = 0.85

# Calculate the probability of cancer given coffee
p_cancer_given_coffee_drinker = \
(p_coffee_drinker_given_cancer * p_cancer) / p_coffee_drinker

# Prints 0.00653846
print("Probability of cancer given coffee: {:.2f}%".format(p_cancer_given_coffee_drinker * 100))
```
