# Number Theory

1. Natural Number - 1,2,3....
2. Whole Number - 0,1,2,3,...
3. Integers - ...-3,-2,-1,0,1,2,3,...
4. Rational Number - all finite decimals & integers - 2/3
5. Irrational Number - Can't expressed as fraction, pi, root of 2 and Eulers number. These numbers have an infinite number of decimal digits, such as 3.141592653.....

Real Number - Rational + Irrational
Complex and Imaginary Number - i - square root of negative number. - Used in matrix decomposition.
